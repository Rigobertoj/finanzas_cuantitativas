```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Download GOOG
ticker = "GOOG"

data = yf.download(
    ticker,
    start="2001-01-01",
    end="2025-09-01",
    interval="1d",
    auto_adjust=False
)

# Nos quedamos solo con el precio ajustado
data = data[["Adj Close"]].rename(columns={"Adj Close": "adjclose"})

# Grfica de precios
data.adjclose.plot(title=f"{ticker} Adjusted Close")
plt.show()

# Histograma de precios
data.adjclose.hist(bins=50)
plt.show()

# Retornos
data["R"] = data.adjclose.pct_change()
data["r"] = np.log(data.adjclose) - np.log(data.adjclose.shift(1))

# Eliminar NaN
data = data.dropna()
```
