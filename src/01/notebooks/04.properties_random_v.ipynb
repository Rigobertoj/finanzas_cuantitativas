```python
import pandas as pd
import numpy as np
```

```python
#r = alpha + betha((rm - rf)) + rf

#B ~ U(0.5, 1)

#alpha ~ N(0.1, 0.15^2)

rf = 0.03
rm = 0.13

mean_beta = (0.5 + 1) / 2
mean_alpha = 0.1

#Expected Value
R = mean_alpha + mean_beta * (rm - rf) + rf
print("Expected Return: ", R)
```

```python
var_alpha = 0.15 ** 2
var_beta = ((1 - 0.5) ** 2) / 12

var_R = var_alpha + ((rm - rf) ** 2) * var_beta + 0 #2Cov(alpha, beta) = 0

print("Variance of Return: ", var_R)
```

```python
print("Standard Deviation of Return: ", var_R ** 0.5)
```

```python
n = 10000

alpha = pd.DataFrame(np.random.normal(0.1, 0.15, n), columns=['alpha'])

beta = pd.DataFrame(np.random.uniform(0.5, 1, n), columns=['beta'])
```
